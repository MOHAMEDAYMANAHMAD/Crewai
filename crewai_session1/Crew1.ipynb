```python
from crewai import LLM, Agent, Task, Process, Crew
```

```python
from dotenv import load_dotenv
```

```python

llm = LLM(
    model="groq/llama-3.1-8b-instant",
    temperature=0
)
```

```python
author_agent = Agent(
    role="Author",
    goal="generate stories given some details about it",
    backstory="""You are a mysterious novelist deeply influenced by Ahmed Khalid Tawfiq's 'Ma Waraa Al Tabiaa'. 
            Your writing is dark, full of suspense, and uses philosophical metaphors. 
            You focus on the psychological state of characters and the eerie atmosphere.""",
    llm=llm,
    verbose=True
)
write_task = Task(
    description="Write a mysterious 10-line story about {person_name} and {subject} in the style of Ahmed Khalid Tawfiq.",
    expected_output="A dark, suspenseful short story.",
    agent=author_agent
)
```

```python
summrizer_agent = Agent(
    role="Summarizer ",
    goal="summarize stories professionally",
    backstory="An experienced American author with 5 years of experience in editing and condensing narratives.",
    llm=llm,
    verbose=True
)
summary_task = Task(
    description="""CRITICAL INSTRUCTION: Read the story provided in the context. 
    Your ONLY job is to write a 4-line summary. 
    DO NOT repeat the story word-for-word. 
    Use your own words as a Chief Editor.""",
    expected_output="A very short, 4-line summary that is completely different in wording from the story.",
    agent=summrizer_agent,
    context=[write_task]
)
```

```python
crew = Crew(
    agents=[
        author_agent,
        summrizer_agent
        ],
    tasks=[
        write_task,
        summary_task
        ],
    process = Process.sequential
)
```

```python
result = crew.kickoff(inputs={
    'person_name': 'Ali',
    'subject': 'his career'
    })
```
